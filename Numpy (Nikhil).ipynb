```python
# Numpy Iterations
# https://docs.scipy.org/doc/numpy/reference/arrays.nditer.html

a = np.arange(6).reshape(2,3)
print("a:\n",a, "\n")
print("a.T:\n",a.T, "\n")

print("x in nditer(a):")
for x in np.nditer(a):
    print(x, end=' ')

print("\n\nx in nditer(a.T):")
for x in np.nditer(a.T):
    print(x, end=' ')

print("\n")

print("\n\nx in nditer(a.T.copy(order='C')):")
for x in np.nditer(a.T.copy(order='C')):
    print(x, end=' ')
    
'''The elements of both a and a.T get traversed in the same order, namely the order they are stored in memory,
whereas the elements of a.T.copy(order=C) get visited in a different order because
they have been put into a different memory layout.'''
print()
print("\nRow-wise:")
for x in np.nditer(a, order='C'): #C-style. Row-wise
    print(x, end=' ')
    
print()
print("Column-wise:")
for x in np.nditer(a, order='F'): #Fortran-style. Column-wise
    print(x, end=' ')
```

```python
# using readwrite option to change each element of array.

for x in np.nditer(a,op_flags=['readwrite']):
    x[...] = x *2
    
print(a)
```

```python
# External Loops
# See difference in the 'chunk' size returned
print(a,"\n\n")
for x in np.nditer(a, flags=['external_loop']):
    print(x, end=' ')
print()
for x in np.nditer(a, flags=['external_loop'], order='F'):
    print(x, end=' ')
print()
for x in np.nditer(a, flags=['external_loop'], order='C'):
    print(x, end=' ')
```

```python
# Multi-index

a = np.arange(6,12).reshape(2,3)
print(a,"\n")
it = np.nditer(a, flags=['f_index'])

while not it.finished:
    print("%d <%d>" % (it[0], it.index), end=' ')
    it.iternext()
    
# You can see that each element has a single index and has been flattened out
```

```python
it = np.nditer(a, flags=['multi_index'])

while not it.finished:
    print("%d <%s>" % (it[0], it.multi_index), end=' ')
    it.iternext()

# You can see that each element has multiple index and has NOT been flattened out

# it = np.nditer(a, flags=['multi_index', 'external_loop']) --> This cannot be used with EXTERNAL_LOOP.
# Error will be thrown: Iterator flag EXTERNAL_LOOP cannot be used if an index or multi-index is being tracked
print()

# Same results with a FOR loop
it = np.nditer(a, flags=['multi_index'])
for x in it:
    print("%d <%s>" % (it[0], it.multi_index), end=' ')
```

```python
# Buffered

a = np.arange(6).reshape(2,3)
print(a)
print()
for x in np.nditer(a, flags=['external_loop'], order='C'):
    print(x, end=' ')

print()

for x in np.nditer(a, flags=['external_loop','buffered'], order='C'):
    print(x, end=' ')

print("\n-----------------------------------------------------------------------------")
# See difference in buffered output: chunk size has increased

for x in np.nditer(a, flags=['external_loop'], order='F'):
    print(x, end=' ')

print()

for x in np.nditer(a, flags=['external_loop','buffered'], order='F'):
    print(x, end=' ')
```
